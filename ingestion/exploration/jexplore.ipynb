```python
from nba_api.stats.endpoints import playercareerstats, draftcombinestats
import pandas as pd
```

```python
data  = draftcombinestats.DraftCombineStats(league_id='00',season_all_time="2024-25")

pd_data = data.draft_combine_stats.get_data_frame()
pd_data.head()
```

```python
missing = pd_data.isnull().sum()
missing_pct = (pd_data.isnull().sum() / len(pd_data)) * 100

# View as a summary
missing_df = pd.DataFrame({
    'column': pd_data.columns,
    'missing_count': missing.values,
    'missing_pct': missing_pct.values
}).sort_values('missing_pct', ascending=False)

print(missing_df)
```

```python
# Filter columns with missing_pct <= 50%
cols_to_keep = missing_df[missing_df['missing_pct'] < 50]['column'].tolist()
pd_data_clean = pd_data[cols_to_keep]

print(f"Kept {len(cols_to_keep)} columns, dropped {len(pd_data.columns) - len(cols_to_keep)} columns")
print(f"Original shape: {pd_data.shape}")
print(f"New shape: {pd_data_clean.shape}")
```
